```python
def findmat (thisdf):
    for record in thisdf.iterrows():
        img_path = record[1]['Path']
        img_c = cv2.imread(img_path)
        try:
            if (img_c.shape[0] == img.shape[0]) and (img_c.shape[1] == img.shape[1]):
                img_dif = np.sum(img-img_c)
                path = 'None'
                if img_dif == 0:
                    print(img_path)
                    path = img_path
                return path
        except:
            path = 'None'
            print('this is none')
```
